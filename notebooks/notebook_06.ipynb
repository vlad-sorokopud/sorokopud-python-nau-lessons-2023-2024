```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
im = cv2.imread('img2.jpg')
im = cv2.resize(im, (600, 600))
plt.imshow(cv2.cvtColor(im, cv2.COLOR_BGR2RGB))
```

```python
img = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_GRAY2RGB))
```

```python
i1 = np.float32(img) # change from uint8 to gloat32
d1 = cv2.cornerHarris(i1, 2, 3, 0.02)
d1 = cv2.dilate(d1, None)

o1 = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
o1[d1 > 0.001 * d1.max()] = [0, 0, 255]
plt.imshow(cv2.cvtColor(o1, cv2.COLOR_BGR2RGB))
```

```python
i2 = cv2.threshold(img, 245, 255, cv2.THRESH_BINARY)[1]
i2 = np.float32(i2) # change from uint8 to gloat32
d2 = cv2.cornerHarris(i2, 2, 3, 0.01)
d2 = cv2.dilate(d1, None)

o2 = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
o2[d2 > 0.00001 * d2.max()] = [0, 0, 255]
plt.imshow(cv2.cvtColor(o2, cv2.COLOR_BGR2RGB))
```

```python
plt.figure(figsize=(6, 4))

plt.subplot(121)
plt.imshow(cv2.cvtColor(o1, cv2.COLOR_BGR2RGB))

plt.subplot(122)
plt.imshow(cv2.cvtColor(o2, cv2.COLOR_BGR2RGB))

plt.show()
```
