```python
%matplotlib inline
import numpy as np
import random
import matplotlib.pyplot as plt

np.random.seed(314156)
```

```python
xa1 = np.array([x + random.randint(1,5) for x in range(100)])
xa1.sort()
xa1
```

```python
ya1 = np.random.rand(100)
ya1
```

```python
plt.plot(xa1)
plt.show()
```

```python
plt.plot(ya1)
plt.show()
```

```python
plt.plot(xa1, ya1)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(xa1, ya1)

#yt = [0.6 for x in range(100)]
#plt.plot(xa1, yt)
ax.axhline(y=0.6, color='r', linestyle=':')

ya11 = [y for y in ya1 if y > 0.6]
ya1i = [i for i, y in enumerate(ya1) if y > 0.6]
xa11 = [xa1[ya1i[i]] for i in range(len(ya1i))]
plt.plot(xa11, ya11, 'r^')


plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.figure(figsize=(12, 2))

plt.subplot(121)
plt.plot(xa1, ya1, color="red")

plt.subplot(122)
plt.plot(xa1, ya1)
yt = [0.6 for x in range(100)]
plt.plot(xa1, yt)

plt.show()
```

```python
plt.scatter(xa1, ya1, color="#72C3D7")

ya11 = [y for y in ya1 if y > 0.6]
ya1i = [i for i, y in enumerate(ya1) if y > 0.6]
xa11 = [xa1[ya1i[i]] for i in range(len(ya1i))]
plt.scatter(xa11, ya11, color='red')

plt.show()
```
